```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime
#Colorama
from colorama import Fore, Back, Style # For text colors
y_= Fore.CYAN
m_= Fore.WHITE
```

# Bitcoin histerical data

```python
# The Ticker module, which allows you to access ticker data in a more Pythonic way:
# history = get historical market data
# period= max will get us data from start of the bit coin to today (2022-09-23) 
# interval = 1d will allow us to get data on daily basis
btc = yf.Ticker("BTC-USD").history(period = "max", interval="1d") 
btc.info()
```

```python
btc.head()
```

```python
print(f"{m_}Total records:{y_}{btc.shape}\n")
print(f"{m_}Data types of data columns: \n{y_}{btc.dtypes}")
```

# Exporting Data of All 10 Types of Cryptocurrencies Into Separate CSV Files

```python
Path = 'F:\\Ml files\\crypto price prediction\\data\\'
```

```python
# to_csv function can take a datafram and file name as input and exports the data into the data folder 
def to_csv(dataframe, file_name):
    return dataframe.to_csv(Path+file_name, index = True)
```

```python
to_csv(btc, 'bitcoin_data.csv')
to_csv(eth, 'etherium_data.csv')
to_csv(ltc, 'litcoin_data.csv')
# to_csv(xrp, 'ripple_data.csv')
# to_csv(tether, 'tether_data.csv')
# to_csv(USD_Coin, 'USD_Coin_data.csv')
to_csv(BNB_Coin, 'BNB_Coin_data.csv')
# to_csv(ADA_Coin, 'Cardano_data.csv')
# to_csv(DOGE_Coin, 'doge_data.csv')
# to_csv(DOT_Coin, 'polkadot_data.csv')
```
